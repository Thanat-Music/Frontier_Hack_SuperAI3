```python
import numpy as np
import pandas as pd
import glob
from PIL import Image
```

```python
input_path = '/lustrefs/disk/project/lt900008-ai2307/train'
output_path = '/lustrefs/disk/project/lt900008-ai2307/train-lable/'
data_path = glob.glob(input_path + "/*.npy")
```

```python
for i in range(len(data_path)):
    lable_path = data_path[i][:-4] + "_labels.csv"
    df = pd.read_csv(lable_path)
    frames = np.load(data_path[i])
    for j, frame in enumerate(frames):
        im = Image.fromarray(frame)
        im = im.convert("L")
        im.save(output_path + df.labels[j]+"/" + str(i)+ ":" + str(j) + ".png") 
    print(i)
print('finish')
```
